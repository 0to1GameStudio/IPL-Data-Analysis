```python
#import pandas to read, manipulate data from source.
import pandas as pd
```

```python
#First step is to read the .csv file for further analysis. Using pandas library.
df = pd.read_csv("deliveries.csv")
df.head()
```

```python
#displayed unique number of rows ['match_id'] column.
df['match_id'].sort_values(ascending= False).unique()  
#df['match_id'].unique()
```

```python
#Describing or explaining data structure of dataset. 
#Analyzing of what datatypes are been used and whether they aligned correctly to the specified columns.
df.info()
```

```python
# Creating and displaying matches played by team 'sunrises_hyd' of match id = 1426312.
sunrises_hyd = df.query('match_id == 1426312')
sunrises_hyd
```

```python
#ccreating a dataframe that displays and uses further analysis of each and another player that matches the condition :
# inning == 1 and team should be 'sunrises_hyd'.
first_innings_batting = sunrises_hyd.query('inning == 1')
first_innings_batting
```

```python
#displaying the rows of player = 'AK Markram' in IPL dataset.
first_innings_batting[first_innings_batting['batter'] == 'AK Markram'].head()
```

```python
#Creating a Temporary dataframe for batters scorecard.
filtered_df = first_innings_batting['extras_type'].isnull()
# Group by batter and aggregate
scorecard_df = (
    first_innings_batting[filtered_df].groupby('batter', as_index=False)
    .agg(
        runs=('batsman_runs', 'sum'),
        balls=('ball', 'count'),
        fours=('batsman_runs', lambda x: (x == 4).sum()),
        sixes=('batsman_runs', lambda x: (x == 6).sum())
    )
)

# Calculate Strike Rate
scorecard_df['S/R'] = (scorecard_df['runs'] * 100 / scorecard_df['balls']).round(2)
scorecard_df.sort_values(by = 'batter', ascending= True)
```

```python
#concatination of two rows using 'over' and 'balls' to create another dataframe that displays batsman's order.
first_innings_batting['over-ball'] = (first_innings_batting['over'].astype(str) + '.' + first_innings_batting['ball'].astype(str)).astype(float)

# Calculate the minimum of 'over-ball' for each batter by grouping data column 'batter'.
batsman_order = first_innings_batting.groupby('batter').agg(
    order=('over-ball', 'min')
).reset_index()
#created a dataframe that shows or displays the batsman's order to bat in the field or match.
ordered_batsman = batsman_order.sort_values(by='order',ascending=True)
```

```python
#checking if the dataframe displays correct rows. 
# After creating from the above dataframe that displays player played number of balls per over.
ordered_batsman
```

```python
#Creating dataframe for scorecard of batters.
batter_result_df = pd.merge(scorecard_df,ordered_batsman, on='batter', how='inner').sort_values(by='order',ascending=True)
batter_result_df['order'] = batter_result_df['order'].astype(float)
batter_result_df['sequence_batting_order'] = range(1, len(batter_result_df) + 1)
#batter_result_df.drop(columns=['order'],inplace =True)
batter_result_df
```

```python
# Grouping by the 'bowler' column to create sorecard of bowler.
scorecard_bowler_df = (
   first_innings_batting.groupby('bowler')
    .apply(lambda x: pd.Series({
       'runs_conceded': x['total_runs'].sum(),
        'not_by_bowler': x.loc[(x['extras_type'] == 'legbyes') | (x['extras_type'] == 'byes'), 'extra_runs'].sum(),
       'balls': ((x['extras_type'].isnull()) | (x['extras_type'].isin(['legbyes', 'byes']))).sum(),
       'W': (x['is_wicket'] == 1).sum()
    }))
    .reset_index()
)
scorecard_bowler_df
```

```python
scorecard_bowler_df['O'] = (scorecard_bowler_df['balls'] // 6).astype(str) + "." + (scorecard_bowler_df['balls'] % 6).astype(str)
scorecard_bowler_df['R'] = scorecard_bowler_df['runs_conceded'] - scorecard_bowler_df['not_by_bowler'].fillna(0)
scorecard_bowler_df['Econ'] = (scorecard_bowler_df['runs_conceded'] / (scorecard_bowler_df['balls'] / 6)).round(2)
```

```python
# Select specific columns from created dataframe known as bowler scorecard.
selected_columns = ['bowler', 'O', 'R', 'W', 'Econ']
scorecard_bowler_df = scorecard_bowler_df[selected_columns]
scorecard_bowler_df
```
